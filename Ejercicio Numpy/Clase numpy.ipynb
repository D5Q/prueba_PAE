```python
#multiplicacion punto entre 2 vectores (equivale a la medida de la proyeccion de uno sobre otro)
arr6=np.random.random([6])
arr7=np.array([1,2,3,4,5,6])
print(np.dot(arr6,arr7))
print(np.dot(arr7,arr6))
```

```python
#multiplicacion punto entre matriz y vector (arroja un vector)
arr8=np.random.random([6,6])
arr9=[1,2,3,4,5,6]
np.dot(arr8,arr9)
```

```python
#multiplicacion punto entre 2 matrices (no es producto matricial)
arr10=np.random.random([6,6])
arr11=np.random.random([6,6])
np.dot(arr10,arr11)
```

```python
#multiplicacion matricial
np.matmul(arr10,arr11)
```
